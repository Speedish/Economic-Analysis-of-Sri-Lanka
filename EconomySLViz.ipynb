```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
```

```python
df = pd.read_csv('EconomicSL.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df['Exports_USD'] = df['Exports_USD'].str.replace('$','')
df['Imports_USD'] = df['Imports_USD'].str.replace('$','')
df['Foreign_Direct_Investment_Inflows_USD'] = df['Foreign_Direct_Investment_Inflows_USD'].str.replace('$','')
df['Manufacturing_Output_USD'] = df['Manufacturing_Output_USD'].str.replace('$','')
df['Economic_Growth_GDP'] = df['Economic_Growth_GDP'].str.replace('$','')

df['Exports_USD'] = df['Exports_USD'].str.replace('B','')
df['Imports_USD'] = df['Imports_USD'].str.replace('B','')
df['Foreign_Direct_Investment_Inflows_USD'] = df['Foreign_Direct_Investment_Inflows_USD'].str.replace('B','')
df['Manufacturing_Output_USD'] = df['Manufacturing_Output_USD'].str.replace('B','')
df['Economic_Growth_GDP'] = df['Economic_Growth_GDP'].str.replace('B','')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Exports_USD'] = df['Exports_USD'].astype('float64')
df['Imports_USD'] = df['Imports_USD'].astype('float64')
df['Foreign_Direct_Investment_Inflows_USD'] = df['Foreign_Direct_Investment_Inflows_USD'].astype('float64')
df['Manufacturing_Output_USD'] = df['Manufacturing_Output_USD'].astype('float64')
df['Economic_Growth_GDP'] = df['Economic_Growth_GDP'].astype('float64')
df['Economic_Growth_Rate'] = df['Economic_Growth_Rate'].astype('float64')
```

```python
df["Economic_Growth_Rate"] = df["Economic_Growth_Rate"].multiply(100)
```

```python
df = df.sort_values('Year', ascending=True)
```

```python
import seaborn as sns
sns.set(color_codes=True)
```
